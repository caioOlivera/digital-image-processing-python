```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Usando rmin e rmax
def contrast_stretching(image):
    rmin, rmax = np.min(image), np.max(image)
    s1, s2 = 0, 255
    result = np.piecewise(image,
                         [image < rmin, (rmin <= image) & (image <= rmax), image > rmax],
                         [lambda x: s1 * x / rmin, lambda x: (s2 - s1) / (rmax - rmin) * (x - rmin) + s1, lambda x: (255 - s2) / (255 - rmax) * (x - rmax) + s2])
    return result
```

```python
# Usando m como média
def thresholding(image):
    m = np.mean(image)
    s1, s2 = 0, 255
    result = np.piecewise(image,
                         [image < m, image >= m],
                         [lambda x: s1 * x / m, lambda x: (s2 - s1) / (255 - m) * (x - m) + s1])
    return result
```

```python
input_image = plt.imread('Células.jpeg')
```

```python
# Aplica a transformação de contraste stretching
output_image_contrast = contrast_stretching(input_image)
```

```python
# Aplica a transformação de thresholding
output_image_threshold = thresholding(input_image)
```

```python
# Plota as imagens
plt.figure(figsize=(15, 5))
```

```python
plt.subplot(1, 3, 1)
plt.imshow(input_image, cmap='gray')
plt.title('Imagem Original')
```

```python
plt.subplot(1, 3, 2)
plt.imshow(output_image_contrast, cmap='gray', vmin=0, vmax=255)
plt.title('Contrast Stretching')
```

```python
plt.subplot(1, 3, 3)
plt.imshow(output_image_threshold, cmap='gray', vmin=0, vmax=255)
plt.title('Thresholding')
```
