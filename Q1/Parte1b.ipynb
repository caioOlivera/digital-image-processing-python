```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def intensity_slice(image, threshold, preserve_below=True):
    if preserve_below:
        result = np.where(image <= threshold, image, 0)
    else:
        result = np.where(image >= threshold, image, 0)
    return result
```

```python
input_image = plt.imread('Angiograma.jpeg')
```

```python
# Define os parâmetros para o Fatiamento de Níveis de Intensidade
threshold = 150
```

```python
# Aplica o Fatiamento de Níveis de Intensidade
output_image_preserve_below = intensity_slice(input_image, threshold, preserve_below=True)
output_image_preserve_above = intensity_slice(input_image, threshold, preserve_below=False)
```

```python
# Plota as imagens
plt.figure(figsize=(15, 5))
```

```python
plt.subplot(1, 3, 1)
plt.imshow(input_image, cmap='gray')
plt.title('Angiograma da Aorta')
```

```python
plt.subplot(1, 3, 2)
plt.imshow(output_image_preserve_below, cmap='gray')
plt.title('Preservar Abaixo de {}'.format(threshold))
```

```python
plt.subplot(1, 3, 3)
plt.imshow(output_image_preserve_above, cmap='gray')
plt.title('Preservar Acima de {}'.format(threshold))
```
