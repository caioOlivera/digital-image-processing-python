```python
import numpy as np
import matplotlib.pyplot as plt

def histogram_equalization_example():
    # Definindo os valores fornecidos no exemplo
    rk = np.array([0, 1, 2, 3, 4, 5, 6, 7])
    nk = np.array([790, 1023, 850, 656, 329, 245, 122, 81])
    pr = np.array([0.19, 0.25, 0.21, 0.16, 0.08, 0.06, 0.03, 0.02])

    # Tamanho da imagem
    MN = np.sum(nk)

    # Calculando a função de transformação
    sk = np.round(((len(rk) - 1) * np.cumsum(pr))).astype(int)

    # Mapeando os valores originais para os valores equalizados
    equalized_histogram = sk[rk]

    # Calculando o histograma equalizado normalizado
    ps_sk = np.zeros(len(sk), dtype=float)
    for s in range(len(sk)):
        indices = np.where(equalized_histogram == s)[0]
        ps_sk[s] = np.sum(pr[indices])

    ps_sk /= MN  # Normalizando o histograma equalizado

    # Exibindo os resultados
    plt.figure(figsize=(15, 4))

    plt.subplot(131)
    plt.bar(rk, nk, width=0.7, align='center', alpha=0.7)
    plt.title('Original Histogram')

    plt.subplot(132)
    plt.step(rk, sk, where='post', color='r')
    plt.title('Transformation Function')

    plt.subplot(133)
    plt.bar(range(len(equalized_histogram)), ps_sk, width=0.7, align='center', alpha=0.7)
    plt.title('Equalized Histogram')

    plt.tight_layout()
    plt.show()

# Executando o exemplo
histogram_equalization_example()
```
